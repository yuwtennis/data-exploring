### Setup adapter

```python
from langchain_mcp_adapters.client import MultiServerMCPClient
from langgraph.prebuilt import create_react_agent
import getpass

mcp_servers = {
    "brightdata": {
        "command": "npx",
        "args": ["@brightdata/mcp"],
        "env": {
            "API_TOKEN": "a8a0c8ed-61b0-4b54-8770-4cf1b996fc52"
        },
        "transport": "stdio"
    }
}

client = MultiServerMCPClient(mcp_servers)
```

```python
tools = await client.get_tools()
```

### LLM for agent

```python
from langchain_anthropic import ChatAnthropic 

model_name = "claude-3-5-sonnet-latest"

llm = ChatAnthropic(model=model_name)
```

### Create Agent

```python
agent = create_react_agent(llm, tools)
```

### Try it out

```python
from langchain_core.messages import HumanMessage
from pprint import pprint
import asyncio

content = "Go to nikkei.com, click the 'トップ' button on the navigation bar, extract the contents in the page."

response = await agent.ainvoke({"messages": [HumanMessage(content=content)]})
```

```python
for m in response["messages"]:
    print(m.content)
```
